```python
import matplotlib.pyplot as plt

games = ['Minecraft', 'GTA V', 'Tetris', 'Wii Sports', 'PUBG',
         'Mario Kart 8', 'Super Mario Bros', 'Pokemon Red/Blue',
         'Wii Fit', 'Terraria']
sales = (300, 190, 170, 83, 75, 60, 58, 47, 43, 44)

plt.figure(figsize=(12,9))
plt.bar(games, sales, width=0.6)

plt.xlabel("Video Game")
plt.ylabel("Sales (Millions)")
plt.title("Top 10 Best-Selling Video Games")
plt.xticks(rotation=45)

# Add labels on top of bars
for x, y in zip(games, sales):
    label = "{:.0f}".format(y)
    plt.annotate(label, (x, y), textcoords="offset points",
                 xytext=(0,10), ha='center')

plt.show()
```
