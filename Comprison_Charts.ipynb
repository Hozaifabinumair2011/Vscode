```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('gapminder(2007).csv')
```

```python
data.head()
```

```python
data.info()
```

```python
avg_data = data.groupby('continent')[['population', 'life_exp', 'gdp_cap']].mean()
```

```python
avg_data = avg_data.reset_index()
```

```python
avg_data
```

```python
plt.bar(avg_data['continent'], avg_data['life_exp'], color='teal')
plt.xlabel('Continent')
plt.ylabel('Average life Expectancy')
plt.show()
```

```python
plt.bar(avg_data['continent'],avg_data['gdp_cap'], color='darkred')
plt.xlabel('Continent')
plt.ylabel('Average GDP per capita')
plt.show()
```

```python
sns.countplot(x=data['continent'], palette='winter')
```

```python
bars1 = [12,28,1,8,22]
bars2 = [28,7,16,4,10]

r=[0,1,2,3,4]

# Corrected: The 'names' list should have the same number of elements as 'r'
names = ['A','B','C','D','E']
barwidth = 1

plt.bar(r,bars1,color='blue',edgecolor='white',width=barwidth)
plt.bar(r,bars2,bottom=bars1,color='red',edgecolor='white',width=barwidth)

plt.xticks(r,names,fontweight='bold')
plt.xlabel("group")

plt.show()
```
